```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Bike Sharing Data')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['dteday'], df['cnt'], label="Daily Rentals")
plt.xlabel("Date")
plt.ylabel("Total Rentals")
plt.title("Daily Bike Rentals Over Time")
plt.legend()
plt.show()
```

```python
df =df.drop(["instant"], axis=1)
df =df.drop(["dteday"], axis=1)
df=df.drop(["registered"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('cnt',axis=1)
y=df['cnt']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("Linear Regression RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print("Linear Regression R2:", r2_score(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("Random Forest R2:", r2_score(y_test, y_pred_rf))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(np.sort(y_test.values), label='Actual')
plt.plot(np.sort(y_pred_lr), label='Predicted')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title("LinearRegression")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(np.sort(y_test.values), label='Actual')
plt.plot(np.sort(y_pred_rf), label='Predicted')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title("RandomForestRegressor")
plt.legend()
plt.show()
```
